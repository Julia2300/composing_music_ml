```python
# compute note range
df["note_range"] = df["note_highest"] - df["note_lowest"]
```

```python
# print number of songs over 3 octaves and filter them out in a new dataframe
print(len(df[df["note_range"] > 24]))
df_within_2_octaves = df[df["note_range"] <= 24]
df_within_2_octaves.head()
```

```python
# check number of songs which begin and end in each octave
octaves = []
for octave in range(2,8):
    octave_start = (octave+1)*12
    octave_end = ((octave+2)*12)-1

    count_begin_end = len(df_within_2_octaves[df_within_2_octaves["note_lowest"] < octave_end])
    count_begin_start = len(df_within_2_octaves[df_within_2_octaves["note_lowest"] < octave_start])

    count_fin_end = len(df_within_2_octaves[df_within_2_octaves["note_highest"] > octave_end])
    count_fin_start = len(df_within_2_octaves[df_within_2_octaves["note_highest"] > octave_start])

    octaves.append([octave, count_begin_end - count_begin_start, count_fin_start - count_fin_end])
pd.DataFrame(octaves, columns= ["octave", "begin", "finish"])
```

```python
def transpose_to_octave_4_to_6(score):
    """
    transposes music score to start in octave 4

    :param score: MIDI file in music21 score format
    :return: transposed MIDI file in music21 score format
    """
    lowest = midi_score.parts[0].pitches[0]
    for part in midi_score.parts:
        for note in part.pitches:
            if note < lowest:
                lowest = note

    transposing_octaves = 4 - lowest.octave
    for part in midi_score.parts:
        for note in part.pitches:
            note.transpose(transposing_octaves*12, inPlace=True)

    return score
```

```python
# when paths not exist, create directories
if not os.path.exists(f"{PATH_TRANSPOSED}/c)_transposed_octave/{dir}"):
    os.makedirs(f"{PATH_TRANSPOSED}/c)_transposed_octave/{dir}")

skipped = {"within_3_or_more_octaves": []}

# transpose octave
for f in tqdm(transposed_files):
    # skip not MIDI files and already transposed ones
    if f[-4:] != ".mid" or os.path.exists(f"{PATH_TRANSPOSED}/c)_transposed_octave/{dir}/{f}"):
        continue
    try:
        note_range = df[df["name"] == f]["note_range"].values[0]
    except:
        continue
    # skip songs over more than 2 octaves
    if note_range > 24:
        skipped["within_3_or_more_octaves"].append(f)
    # convert other songs to octave 4 to 6
    else:
        try:
            midi_score = converter.parse(f"{PATH_TRANSPOSED}/b)_transposed_key/{dir}/{f}")
            midi_score = transpose_to_octave_4_to_6(midi_score)
            midi_score.write("midi", fp=f"{PATH_TRANSPOSED}/c)_transposed_octave/{dir}/{f}")
        except:
            continue

skipped
```
